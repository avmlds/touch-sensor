```python
import pandas as pd
import seaborn as sns
```

```python
d = pd.read_csv("test_data_025_125_20.csv")
d2 = pd.read_csv("test_data_025_125_20_flat_needle_water.csv")
d3 = pd.read_csv("test_data_025_125_20_flat_needle_soap.csv")
d4 = pd.read_csv("test_data_025_125_20_sharp_needle_soap.csv")
d5 = pd.read_csv("test_data_1_5_20_sharp_needle.csv")
d6 = pd.read_csv("test_data_1_5_20_flat_needle_single.csv")
```

```python
sns.set(rc={'figure.figsize':(20,10)})
```

```python
sns.lineplot(data=d[["mkm_passed", "diff_pa_aver"]], x="mkm_passed", y="diff_pa_aver")
```

```python
import numpy as np
```

```python
d.groupby("mkm_passed").agg(
    {
        "diff_pa_aver": [np.mean, np.median, np.var, np.std]
    }
).reset_index().loc[90:110]
```

```python
sns.lineplot(data=d2[["mkm_passed", "diff_pa_aver"]], x="mkm_passed", y="diff_pa_aver")
```

```python
df = d[["epoch", "mkm_passed", "diff_pa_aver", "time"]]
df1 = d2[["epoch", "mkm_passed", "diff_pa_aver", "time"]]
```

```python
df.columns = [" ", ", ", " ", ""]
df1.columns = [" ", ", ", " ", ""]
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
```

```python
for n, i in df1.groupby(" "):
    plt.scatter(i.loc[(i[", "] > 6000) & (i[", "] < 8500)][", "], i.loc[(i[", "] > 6000) & (i[", "] < 8500)][" "], label=n)

plt.legend(title=" ")
plt.xlabel("  ")
plt.ylabel(", ")
plt.title("         ")
```
