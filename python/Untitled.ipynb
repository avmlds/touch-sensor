```python
import pandas as pd
```

```python
df = pd.read_csv("data2.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df = df.drop(df.loc[df["flow"] == ' '].index, axis=0)
df["flow"] = df["flow"].astype(int)
```

```python
df["step_1"] = df["step"].copy()
```

```python
j = 0
for n, i in enumerate(df["step"]):
    df.loc[n, "step_1"] = j
    j += 1
    if i == 0:
        j = 0
```

```python
df[["valve", "flow", "step", "touch", "step_1"]] = df[["valve", "flow", "step", "touch", "step_1"]].astype(int)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 10))
# 204 looks good
data = df.loc[(df["valve"] == 193) & (df["direction"] == "f")]
sns.lineplot(x=data["step_1"], y=data["flow"], ax=ax)
```
