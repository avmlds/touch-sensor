```python
import pandas as pd
```

```python
df = pd.read_csv("data.txt")
```

```python
df.head()
```

```python
df.columns = ["valve", "flow", "direction", "counter"]
# df = df.apply(lambda x: x.replace('"', "")).copy()
```

```python
#df["valve"] = df["valve"].apply(lambda x: int(x.replace('"', "").strip()) if x.replace('"', "").strip() != "" else 0).astype(int)
df["flow"] = df["flow"].apply(lambda x: int(x.replace('"', "").strip()) if x.replace('"', "").strip() != "" else 0).astype(int)
df["direction"] = df["direction"].apply(lambda x: 1 if "f" in x else 0).astype(int)
df["counter"] = df["counter"].apply(lambda x: int(x.replace('"', "").strip())).astype(int)
```
