```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
df = df.drop(df.loc[df["flow"] == ' '].index, axis=0)
df["flow"] = df["flow"].astype(int)
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(20, 10))
# 204 looks good
data = df.loc[(df["valve"] == 204) & (df["direction"] == "f") & (df["step"] < 14)]
sns.lineplot(x=data["step"], y=data["flow"], ax=ax)
```
